```python
import pandas as pd
```

```python
cacc = pd.read_csv("C:\\games\\\Edulyt project\\customer_acquisition.csv")
cacc
spend=pd.read_csv("C:\\games\\\Edulyt project\\spend.csv",parse_dates=["Month"])
spend
repay=pd.read_csv("C:\\games\\\Edulyt project\\repayme_new.csv",parse_dates=["Month"])
repay
cacc
```

# Data Cleaning

```python
mean_age=cacc.loc[cacc["Age"]>=18,"Age"].mean()
mean_age=round(mean_age,4)
mean_age
```

# Meaningful treatment of age

```python
cacc.loc[cacc["Age"]<18,"Age"]=mean_age
cacc.to_csv('Data_Cleaning_1.csv')
cacc
```

# Credit of 2 %

```python
##spend["Month"]=pd.to_datetime(spend["Month"])
spend["mon"]=spend["Month"].dt.month
spend["year"]=spend["Month"].dt.year
spend

repay["mon"]=repay["Month"].dt.month
repay["year"]=repay["Month"].dt.year
repay
```

```python
sum_spend=spend.groupby(["Costomer","year","mon"])["Amount"].sum().reset_index(name ='Total spend')
sum_spend

sum_repay=repay.groupby(["Costomer","year","mon"])["Amount"].sum().reset_index(name ='Total repayment')
sum_repay
```

```python
spend_repay=sum_spend.merge(sum_repay,on=["Costomer","year","mon"],how="outer")
spend_repay.loc[pd.isna(spend_repay["Total spend"]),"Total spend"]=0
spend_repay.loc[pd.isna(spend_repay["Total repayment"]),"Total repayment"]=0   
spend_repay["credit"]=0
spend_repay
```

```python
row=len(spend_repay.index)
spend_repay["interest"]=0
for i in spend_repay.index:
     if(spend_repay["Total repayment"][i]>spend_repay["Total spend"][i]):
            spend_repay.loc[spend_repay.index==i,"credit"]=(spend_repay["Total repayment"][i]-spend_repay["Total spend"][i])*0.02
            if(i!=row-1 and spend_repay["Costomer"][i+1]==spend_repay["Costomer"][i]):
                   spend_repay.loc[spend_repay.index==i+1,"Total spend"]+=spend_repay["credit"][i]
     elif(spend_repay["Total spend"][i]>spend_repay["Total repayment"][i]):
            spend_repay.loc[spend_repay.index==i,"interest"]=(spend_repay["Total spend"][i]-spend_repay["Total repayment"][i])*0.029
            if(i!=row-1 and spend_repay["Costomer"][i+1]==spend_repay["Costomer"][i]):
                   spend_repay.loc[spend_repay.index==i+1,"Total repayment"]+=spend_repay["interest"][i]              
                    
                    
spend_repay.to_csv('Data_Cleaning_2.csv')
spend_repay
```

# Category spending most money

```python
spend["Category"]=" "
for i in spend.index:
    if(spend["Type"][i]=="AIR TICKET" or spend["Type"][i]=="AUTO" or spend["Type"][i]=="BIKE" or spend["Type"][i]=="BUS TICKET" or spend["Type"][i]=="CAR" or spend["Type"][i]=="PETRO" or spend["Type"][i]=="TRAIN TICKET"):
        spend.loc[spend.index==i,"Category"]="Travel"
    elif(spend["Type"][i]=="FOOD" or spend["Type"][i]=="RENTAL"):
        spend.loc[spend.index==i,"Category"]="Accomodation"
    elif(spend["Type"][i]=="CAMERA" or spend["Type"][i]=="MOVIE TICKET"):
        spend.loc[spend.index==i,"Category"]="Entertainment"
    else:
        spend.loc[spend.index==i,"Category"]="Accessories"
category_spend=spend.groupby(["Category","year","mon"])["Amount"].sum().reset_index(name="Total Spend")
category_spend=category_spend.sort_values(by=["Total Spend"],ascending=False)
category_spend.to_csv('Task_7.csv')
category_spend
```
